```python
import scipy.io
import h5py
data = scipy.io.loadmat('plot_savespace.mat')
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
f = h5py.File('plot_savespace_v2.mat','r')
mu = f.get('data/mu')
mu = np.array(mu) # For converting to a NumPy array
```

```python
mu = data['mu']
mu = mu.reshape((100, 1))
mu = mu.flatten()

E_comp_vs_mu_rail = data['E_comp_vs_mu_rail']
E_comp_vs_mu_rail = E_comp_vs_mu_rail.reshape((100, 1))
E_comp_vs_mu_rail = E_comp_vs_mu_rail.flatten()

E_comp_vs_mu_bench = data['E_comp_vs_mu_bench']
E_comp_vs_mu_bench = E_comp_vs_mu_bench.reshape((100, 1))
E_comp_vs_mu_bench = E_comp_vs_mu_bench.flatten()
```

```python
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
ax.plot(mu, E_comp_vs_mu_rail, linewidth=2, label='rail')
ax.plot(mu, E_comp_vs_mu_bench, linestyle='dashed', linewidth=2, label='benchmark')
ax.fill_between(mu, E_comp_vs_mu_bench, E_comp_vs_mu_rail, where=(E_comp_vs_mu_bench > E_comp_vs_mu_rail), color='darkorange', alpha=0.2)
ax.fill_between(mu, E_comp_vs_mu_bench, E_comp_vs_mu_rail, where=(E_comp_vs_mu_bench < E_comp_vs_mu_rail), color='dodgerblue', alpha=0.2)
#ax.plot(x, y)
ax.legend()

ax.set_xlabel('$\mu$', fontsize=16)
ax.set_ylabel('$\mathcal{E}$ (J)', fontsize=16)

ax.set_xticks([0, 0.05, 0.1], labels=['0', '0.05', '0.1'], fontsize=16)
ax.set_yticks([250, 500, 750], labels=[250, 500, 750], fontsize=16)


plt.savefig('E_vs_mu.pdf')

plt.show()
```
